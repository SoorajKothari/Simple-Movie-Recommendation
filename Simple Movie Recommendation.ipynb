```python
credit = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

## Merging Movies Dataset with Credits Dataset with respect to ID as ID is comman in both dataset

```python
credit_column_new = credit.rename(index=str,columns={"movie_id":"id"})
movies_df_merge = movies.merge(credit_column_new,on='id')
```

```python
movies_df_merge.head()
```

# Dropping unless column to simplify our dataset

```python
movies_new = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_new.head()
```

```python
movies_new.info()
```

# There is average rating of every movie in dataset with column name vote_average so we i have used Average weight as sample recommendation system Solution

```python
movies_new['vote_average'].head()
```

# Weigted Rating Formula
## W = (R*v + c*m) / (v+m)
###### Here W = Weight Rating
###### Here R =  vote average
###### Here v = vote count
###### Here m = minimum votes required to be listed
###### Here c = mean votes accross dataset

```python
V = movies_new['vote_count']
R = movies_new['vote_average']
C = movies_new['vote_average'].mean()
M = movies_new['vote_count'].quantile(0.70)
```

```python
movies_new['weighted_avg'] = ((R*V)+ (C*M))/ (V+M)
movies_new['weighted_avg'].head()
```

```python
## Sorting weighted_avg in ascending order
movies_new_sorted = movies_new.sort_values('weighted_avg',ascending=False)
movies_new_sorted[['original_title', 'vote_count', 'vote_average', 'weighted_avg', 'popularity']].head()
```

```python
## Yet i have used vote count and vote avg only but popularity can also b included to get best prediction
## as vote count , vote avg and popularity of movie are all considered mostly.

weight_average=movies_new_sorted.sort_values('weighted_avg',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_avg'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.title('Best Movies by average votes with popularity consideration', weight='bold')
plt.xlabel('Weighted Avg Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

```python
popularity=movies_new_sorted.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes with popularity and weight avg', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

```python
#i have scaled weighted avg and popularity betweem 0 to 1 because as we can weighted_avg has values in 0-10
# while popularity has values in 100s so below is scaling between 0-1 

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_new[['weighted_avg','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_avg','popularity'])
movie_normalized_df.head()
```

```python
## inserting into old dataset the normalized values
movies_new[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
movies_new.head()
```

```python
## since i have tried to predict on 2 features weighted_avg and popularity so taking 50% of both individually and trying to
## get 1 feature which i ev named as score
movies_new['score'] = movies_new['normalized_weight_average'] * 0.5 + movies_new['normalized_popularity'] * 0.5
movies_scored_df = movies_new.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head()
```

```python
scored_df = movies_new.sort_values('score', ascending=False)
plt.figure(figsize=(16,6))
ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```
